```python
import scipy.io.wavfile
import scipy.signal
import numpy as np
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adadelta, Adam
from tensorflow.keras.losses import sparse_categorical_crossentropy #TODO what to import here?
from tensorflow.keras.regularizers import l1
from tensorflow.keras.backend import clear_session
import tensorflow as tf
from tensorflow.keras.layers import Dropout

def WavFileToArray(filename):
    fs, wave = scipy.io.wavfile.read(filename)
```

```python
from sklearn.model_selection import train_test_split


def WavToSplittedArray(filename="", window_size = 100):
    fs, wave = scipy.io.wavfile.read(filename) 
    windows_count = int(len(wave)/window_size)

    kimenet = []
    for i in range(windows_count):
      buffer = []
      for x in range(window_size):
        buffer.append(wave[i*window_size+x])
      kimenet.append(buffer)
    return kimenet


ablakmeret = 500

alapjarat_train = WavToSplittedArray("Billie Eilish - No Time To Die.wav", ablakmeret)
alapjarat_label = [0] * len(alapjarat_train)

gyertya1lenn_train = WavToSplittedArray("Bonnie Tyler - Total Eclipse of the Heart.wav", ablakmeret)
gyertya1lenn_label = [1] * len(gyertya1lenn_train)

gyertya2lenn_train = WavToSplittedArray("Keep Holding On.wav", ablakmeret)
gyertya2lenn_label = [2] * len(gyertya2lenn_train)


x_train_whole = alapjarat_train
y_train_whole = alapjarat_label

x_train_whole.extend(gyertya1lenn_train)
y_train_whole.extend(gyertya1lenn_label)

x_train_whole.extend(gyertya2lenn_train)
y_train_whole.extend(gyertya2lenn_label)


x_train, x_test_valid, y_train, y_test_valid = train_test_split(x_train_whole, y_train_whole, test_size=0.2, random_state=None)
x_test, x_valid, y_test, y_valid = train_test_split(x_test_valid, y_test_valid, test_size=0.5, random_state=None)

x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
x_valid = np.array(x_valid)
y_test = np.array(y_test)
y_valid = np.array(y_valid)

print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(x_valid))
print(len(y_test))
print(len(y_valid))
```

```python

hidden_layer_1_size = 50
hidden_layer_2_size = 30


tf.compat.v1.reset_default_graph() # It's good practice to clea and reset everything
clear_session            # even using Keras


# Model
#######

x = Input(shape=(ablakmeret,))

# Hidden layers


hidden_layer_1 = Dense(units=hidden_layer_1_size, 
                        activation="relu",
                        kernel_regularizer="l1")(x)

d1_layer = Dropout(.2)(hidden_layer_1)
    
hidden_layer_2 = Dense(units=hidden_layer_2_size, 
                        activation="relu",
                        kernel_regularizer="l1")(d1_layer)

d2_layer = Dropout(.2)(hidden_layer_2)
                        
# Softmax 

predictions = Dense(3, activation="softmax")(d2_layer) #TODO: What activation?

# Full model

model = Model(inputs=x, outputs=predictions)

model.summary()
```

```python
# Loss 

loss = sparse_categorical_crossentropy#TODO use the imported loss!

# Optimizer

optimizer = Adam() # A "smart" SGD variant 
 
# Compilation
#############

model.compile(optimizer=optimizer, loss=loss, metrics=["accuracy"]) #TODO what to measure here?
```

```python

history = model.fit(x=x_train, y=y_train,
                    validation_data=(x_valid, y_valid),
                    epochs=200,
                    batch_size=150)
```

```python
import matplotlib.pyplot as plt

def display_history(history):
    plt.plot(history.history['accuracy']) #TODO what to measure here for training?
    plt.plot(history.history['val_accuracy']) #TODO what to measure here for validation?
    plt.title('Model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'valid'], loc='upper left')
    plt.show()
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'valid'], loc='upper left')
    plt.show()
    
display_history(history)
```
