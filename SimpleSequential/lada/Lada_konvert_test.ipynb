```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = pd.read_csv('2lehuzasa_mono.txt', header=None)

# fig = plt.gcf()
# fig.set_size_inches(25.5, 5)


dataMaxSize, qq = x.shape
dataSize = dataMaxSize

if dataSize > dataMaxSize:
    dataSize = dataMaxSize

timeUnit = 1 / 44100
timeSize = dataSize * timeUnit

# sum = 0
# max = 0
# min = 0

# for i in range(0,dataSize):
#   sum += x[0][i]
#   if x[0][i] < min:
#     min = x[0][i]

#   if x[0][i] > max:
#     max = x[0][i]

# print('Sum: ' , str(sum))
# print('min: ', str(min))
# print('max: ', str(max))

zaj = np.random.normal(0,1,1000)
#zaj.shape

# fig = plt.gcf()
# fig.set_size_inches(35.5, 8)
# plt.plot(zaj)
# plt.show()

# #pukligenerátor

# # Generating time data using arange function from numpy
domb = np.arange(0, np.pi, 0.1)

# # Finding amplitude at each time
pukli = 10 * np.sin(domb)

# fig = plt.gcf()
# fig.set_size_inches(35.5, 8)
# plt.plot(pukli)
# plt.show()
# print(len(pukli))

#puklis zaj generátor

pukliStart1 = 300
pukliStart2 = 350
pukliStart3 = 400


for i in range(pukliStart1, pukliStart1 + 30):
    zaj[i] = zaj[i] + pukli[i-pukliStart1]

for i in range(pukliStart2, pukliStart2 + 30):
    zaj[i] += pukli[i-pukliStart2]
    # zaj[i] = pukli[i-pukliStart2]

for i in range(pukliStart3, pukliStart3 + 30):
    zaj[i] = zaj[i] + pukli[i-pukliStart3]

#puklikeresés 1. keresőmátrix tudja hogy pukli ;) valahogy megsejtette.... a galád

matrixSize = 30   # 20 ms = kb 800

matrixProfil = pukli

errorArray = []

for i in range (0, 1000 - matrixSize):
    hiba = 0
    for j in range(0, matrixSize):
        hiba += (zaj[i+j] - pukli[j])**2

    errorArray.append(hiba)
    

    
fig = plt.gcf()
fig.set_size_inches(35.5, 8)
plt.plot(zaj)
plt.show()
```
